```python
import magellan as mg
import pandas as pd
```

```python
A = mg.load_dataset('table_A')
B = mg.load_dataset('table_B')
```

```python
ab = mg.AttrEquivalenceBlocker()
C = ab.block_tables(A, B, 'zipcode', 'zipcode')
```

```python
C
```

```python
D = ab.block_tables(A, B, 'name', 'zipcode', ['name', 'address'], ['name', 'address'])
```

```python
E = mg.combine_block_outputs_via_union([C, D])
```
