```python
import magellan as mg
```

```python
A = mg.load_dataset('table_A')
B = mg.load_dataset('table_B')
```

```python
L = mg.read_csv('lable_ab_correct_labels.csv', ltable=A, rtable=B, 
                foreign_key_ltable='ltable.ID', foreign_key_rtable='rtable.ID')
```

```python
L
```

```python
f = mg.get_features_for_matching(A, B)
```

```python
G = mg.extract_feat_vecs(L, feature_table=f, attrs_after='gold')
```

```python
G
```

```python
train_test = mg.train_test_split(G, random_state=2)
```

```python
train_test['train']
```

```python
train_test['test']
```

```python
lg = mg.LogRegMatcher()
```

```python
lg.fit(table=train_test['train'], exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'], target_attr='gold')
```
